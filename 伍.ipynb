```python
#一、畫圖
import matplotlib.pyplot as pt
x=[1,1,1.5,2,2.5,3,3.5,3.5]
y=[1,3,5,3,3,5,3,1]
pt.plot(x,y)
pt.plot(x,"r--")
pt.plot(y,"b>")
pt.show()
```

```python
#二、隨機畫圖
import matplotlib.pyplot as pt
import numpy as np
x=np.random.rand(1000)
y=np.random.rand(1000)
pt.scatter(x,y)
pt.show()
```

```python
#三、sin wave
import matplotlib.pyplot as pt
import numpy as np
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.cos(x*np.pi/180.0)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.plot(x,y)
pt.plot(x,z)
pt.title("Sin & Cos Wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.show()
```
